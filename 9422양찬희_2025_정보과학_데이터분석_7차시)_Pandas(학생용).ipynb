## pandas란?
판다스는 데이타 분석(Data Analysis)을 위해 널리 사용되는 파이썬 라이브러리로 데이터를 표(DataFrame)형식으로 저장하고 데이터를 관리하는 강력한 도구를 지원합니다.

<img src=https://dandyrilla.github.io/images/2017-08-12/fig0.png width=500>

- 자료구조 : Series(1차원),  DataFrame(2차), Panel(3차원)
- 파일 불러오기 : read_csv, read_excel
- 속성 : describe, dtypes, shape, size, head, tail, index, colnums, astype, to_numeric
- 검색 : 행단위, 열단위, loc, iloc, 조건(==, >=, isin, str.contains('문자'),  &(and), |(or))
- 데이터 도구 : sort_values, is_null(), dropna, fillna
- 관리 : drop(삭제), 열추가, 행추가
- 수학함수들 : sqrt, abs, log, exp, sin, cos, mean, std, max, argmax, cumsum
- 난수 : numpy.random  모듈 내 shuffle, rand, randint, randn
- 기타 : T(역행렬), reshape(구조 변경), merge(병합), axis(축 1:열, 0:행)

📊 Pandas 활용 예시
1. 교육 분야
- 학생들의 시험 점수, 출석률, 과제 제출 현황을 정리하고 분석

- 예: 학급 성적 분석, 성적 변화 추세 시각화, 과목별 평균 비교

2. 비즈니스/마케팅
- 고객 데이터 분석 (연령, 구매 이력, 선호도 등)

- 판매 데이터 분석 (일별/월별 매출, 베스트셀러 상품 찾기)

- 마케팅 캠페인의 효과 분석 (이메일 오픈율, 클릭률 등)

3. 공공 데이터 분석
- 정부에서 제공하는 공공 데이터를 분석하여 사회 문제 해결

- 예: 지역별 교통사고 현황, 미세먼지 농도 변화, 인구통계

4. 스포츠 데이터
- 선수별 경기 기록, 팀별 통계 분석

- 예: KBO, NBA, 월드컵 데이터 분석

5. 사회관계망 서비스(SNS) 분석
- 트위터, 인스타그램 등의 데이터를 수집해서 감정 분석 또는 인기 키워드 추출

- 예: 워드클라우드 만들기, 해시태그 분석, 긍정/부정 리뷰 분류

6. 시간 데이터(Time Series) 분석
- 주가, 기온, 환율, 유튜브 조회수 등 시간이 흐르며 변화하는 데이터 분석

- 예: 주가 변화 예측, 일일 확진자 수 분석, 방문자 수 추세 파악

7. 코로나19 데이터 분석
- 확진자 수, 백신 접종률, 지역별 감염 추세 분석

- 예: 누적 확진자 시각화, 국가별 대응 비교

###우리 동네와 인구 구조와 비슷한 지역들(5개)을 그래프로 나타내기(pandas 사용)

````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family ='NanumGothic')
df = pd.read_csv(csv_url, index_col = 0)

df = df.div(df['총인구수'], axis =0)     # ①-1. 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수']  # ①-2. 총인구수, 연령구간인구수 열 삭제

name = input('원하는 지역의 이름을 입력해주세요 : ')  # ② 지역 이름 입력
a = df.index.str.contains(name)  # ③ 해당 행을 찾아서 해당 지역의 인구 구조를 저장

df2 = df[a]
# 인구비율을 빼고, 제곱값을 모두 더하고, 그 차이가 가장 작은 지역 5곳을 찾아서 꺽은선 그래프로 표현하기
df.loc[np.power(df.sub(df2.iloc[0], axis = 1),2).sum(axis=1).sort_values().index[:5]].T.plot(figsize = (10,7))
plt.show()
````
